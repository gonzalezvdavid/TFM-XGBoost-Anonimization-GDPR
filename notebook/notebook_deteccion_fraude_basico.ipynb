```python
# Detección de Fraude Bancario con Anonimización y Cumplimiento GDPR
# ================================================================
# Este notebook muestra cómo aplicar técnicas de anonimización al conjunto de datos
# PaySim1 para la detección de fraude, manteniendo el cumplimiento del GDPR.

# Importando las funciones del módulo principal
# -----------------------------------------
from eda_anonimizacion_fraud_detection import *
```

```python
# 1. Carga y exploración del conjunto de datos
# ---------------------------------------
# Definiendo la ruta del archivo
ruta_entrada = "dataset/anonimizacion_datos.csv"

# Cargando los datos
df = cargar_datos(ruta_entrada)

# Mostrando la información básica
mostrar_info_basica(df)

# Visualizando valores nulos
visualizar_valores_nulos(df)
```

```python
# 3. Realizando análisis exploratorio detallado
# ---------------------------------------
# Analizando variables numéricas
columnas_numericas = analizar_variables_numericas(df)

# Analizando montos por tipo de transacción
analizar_montos_por_tipo(df)

# Analizando matriz de correlación
analizar_correlaciones(df, columnas_numericas)

# Analizando balances
analizar_balances(df)

# Analizando patrones temporales
analizar_patrones_temporales(df)
```

```python
# 4. Entrenando el modelo de referencia (sin anonimización)
# ------------------------------------------------
print("\n===== ENTRENAMIENTO DE MODELO SIN ANONIMIZACIÓN =====")

# Seleccionando las columnas relevantes para el modelo
columnas_modelo = [
    'type', 'amount', 'oldbalanceOrg', 'newbalanceOrig', 
    'oldbalanceDest', 'newbalanceDest', 'step'
]

# Codificando las variables categóricas
df_procesado = df.copy()
if 'type' in df_procesado.columns:
    le = LabelEncoder()
    df_procesado['type'] = le.fit_transform(df_procesado['type'])

# Preparando los datos para el modelo original
X_train_orig, X_test_orig, y_train_orig, y_test_orig, _ = preparar_datos_para_modelo(
    df_procesado, 
    columnas_modelo=columnas_modelo, 
    test_size=0.3, 
    random_state=42
)

# Entrenando y evaluando el modelo original
modelo_original, precision_orig, sensibilidad_orig, f1_orig = entrenar_y_evaluar_modelo(
    X_train_orig, X_test_orig, y_train_orig, y_test_orig, 
    n_estimators=100, random_state=42
)
```

```python
# 5. Aplicando las técnicas de anonimización
# ---------------------------------
print("\n===== APLICANDO TÉCNICAS DE ANONIMIZACIÓN =====")
# Iniciar proceso de anonimización
df_anonimizado = iniciar_anonimizacion(df)

# Analizando el formato de los IDs
analizar_formato_ids(df_anonimizado)

# Aplicando la seudonimización (hashing SHA-256)
df_anonimizado = anonimizar_ids(df_anonimizado)

# Aplicando la agrupación (k-anonimato)
df_anonimizado = anonimizar_mediante_agrupacion(df_anonimizado)
```

```python
# 6. Verificando cumplimiento de k-anonimato y l-diversidad
# ---------------------------------------------------
# Verificando k-anonimato
k = 10  # Valor de k para k-anonimato
cumple_k = verificar_k_anonimato(df_anonimizado, k=k)

# Verificando l-diversidad
l = 2  # Valor de l para l-diversidad
columnas_agrupacion = [col for col in df_anonimizado.columns if col.endswith('_group')]
if 'type' in df_anonimizado.columns and columnas_agrupacion:
    cumple_l = verificar_l_diversidad(df_anonimizado, columnas_agrupacion, 'type', l=l)
```

```python
# 7. Entrenando el modelo con datos anonimizados
# --------------------------------------
print("\n===== ENTRENAMIENTO DE MODELO CON DATOS ANONIMIZADOS =====")

# Utilizamos las columnas agrupadas en lugar de las originales
columnas_modelo_anon = [
    'type', 'amount_group', 'step_group', 
    'oldbalanceOrg', 'newbalanceOrig', 
    'oldbalanceDest', 'newbalanceDest'
]

# Codificando las variables categóricas
df_anonimizado_procesado = df_anonimizado.copy()

for col in ['type', 'amount_group', 'step_group']:
    if col in df_anonimizado_procesado.columns:
        le = LabelEncoder()
        df_anonimizado_procesado[col] = le.fit_transform(df_anonimizado_procesado[col])

# Preparando los datos para el modelo anonimizado
columnas_disponibles = [col for col in columnas_modelo_anon if col in df_anonimizado_procesado.columns]

X_train_anon, X_test_anon, y_train_anon, y_test_anon, _ = preparar_datos_para_modelo(
    df_anonimizado_procesado, 
    columnas_modelo=columnas_disponibles, 
    test_size=0.3, 
    random_state=42
)

# Entrenando y evaluando el modelo anonimizado
modelo_anonimizado, precision_anon, sensibilidad_anon, f1_anon = entrenar_y_evaluar_modelo(
    X_train_anon, X_test_anon, y_train_anon, y_test_anon, 
    n_estimators=100, random_state=42
)
```

```python
# 8. Comparando el rendimiento de los modelos
# -------------------------------
comparar_modelos(
    (precision_orig, sensibilidad_orig, f1_orig),
    (precision_anon, sensibilidad_anon, f1_anon)
)
```

```python
# 9. Evaluando el cumplimiento del GDPR
# -----------------------------
# Calcular diferencia porcentual en precisión
dif_precision = (precision_orig - precision_anon) / precision_orig * 100

# Evaluando el cumplimiento GDPR
evaluacion_gdpr = evaluar_cumplimiento_gdpr(k, l, dif_precision)
```

```python
# 10. Generando el dashboard de cumplimiento
# ---------------------------------
generar_dashboard_cumplimiento(
    evaluacion_gdpr,
    (precision_orig, sensibilidad_orig, f1_orig),
    (precision_anon, sensibilidad_anon, f1_anon)
)
```
